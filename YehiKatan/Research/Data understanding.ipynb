```python
import scipy.io

# Datei laden
data = scipy.io.loadmat("data/heatcube_0001.mat")

# Alle enthaltenen Variablennamen anzeigen

print(data.keys())
```

```python
HSI = data['HSI']
print("Shape:", HSI.shape)
print("Dtype:", HSI.dtype)
print("Min:", HSI.min(), "Max:", HSI.max())
```

## Bild ansehen

```python
import matplotlib.pyplot as plt

band_index = 48
plt.imshow(HSI[:, :, band_index], cmap='inferno')
plt.title(f"Hyperspektralbild Band {band_index}")
plt.colorbar(label="Intensitt")
plt.axis('off')
plt.show()
```

```python
# wavenumber_ExperimentalScenes

import scipy.io

# Datei laden
# matLib_FullDatabase
data_wn = scipy.io.loadmat("data/wavenumber_ExperimentalScenes.mat")

# Alle enthaltenen Variablennamen anzeigen

print(data_wn.keys())
```

## Spektrum eines bestimmten Pixels anzeigen

```python
y, x = 80, 800 # Pixel (y, x)
spectrum = HSI[y, x, :]


plt.plot(matLibWn, spectrum) # Ohne Wave
plt.title(f"Spektrum bei Pixel ({x}, {y})")
plt.xlabel("Bandindex")
plt.ylabel("Intensitt")
plt.grid(True)
plt.show()
```
